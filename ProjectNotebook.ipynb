# Project Description

In my project, I developed code to automate the process of encrypting and decrypting messages using the Vigenère cipher (I learned about the Vigenère cipher in LIGN 17). 

The Vigenère cipher is a method of encrypting plaintext (ie human readable text, in this case, English) using polyalphabetic substitution. To use this cipher, we build a table that contains all 26 possible Caesar Shift Ciphers, with the first row having no shift (starts with A B C D etc). Each subsequent row is shifted 1 letter to the right (so the second row would start with B C D E etc). In the table, plaintext letters are along the top and the keyword letters are along the left.

To encrypt, you take the message you want to encrypt, pick a keyword to use (it can be anything but is more secure if it is longer and is the same length as the message/never repeats). Write the keyword above the message over and over until the end of the message. The letter in the keyword tells you which row in the table to use to encrypt the corresponding plaintext letter. You then take that letter in plaintext and match it to the letter at the top of the column. Follow the column down to the letter that corresponds with the letter in the keyphrase. Repeat for every letter in the plaintext.

To decrypt, you would do the same process in reverse (you should know what key was used). If you know 2 of the 3 pieces of information (ciphertext, plaintext, keyphrase), you can discover the third. In my project, you are able to find all three pieces of information given two. *When finding the keyword, it returns the keyword repeated to match the length of the message.

NOTE: Vigenère Cipher can be unbreakable if the key is randomly generated, the same length as the ciphertext (so the key never repeats), and only used once. Otherwise, it is vulnerable to being broken using the Babbage-Kasinski method
